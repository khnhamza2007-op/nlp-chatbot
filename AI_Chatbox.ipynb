```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# key words
keywords = [
    "ai ethics",
    "study",
    "honesty",
    "stress"
]

# Convert keywords into numbers
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(keywords)

# Chatbot function
def chatbot(user_input):
    user_input = user_input.lower()

    # IF-ELIF RULE BASED PART
    if user_input in ["hi", "hello"]:
        return "Hello! I am your ethical study chatbot 😊"

    elif "bye" in user_input:
        return "Goodbye! Keep learning ethically 📘"

    # MACHINE LEARNING PART
    user_vec = vectorizer.transform([user_input])
    similarity = cosine_similarity(user_vec, X)
    intent = similarity.argmax()

    # IF-ELIF RESPONSE BASED ON INTENT
    if similarity[0][intent] > 0.3:

        if intent == 0:
            return "AI ethics means using AI responsibly and fairly."

        elif intent == 1:
            return "Study regularly, avoid plagiarism, and be honest."

        elif intent == 2:
            return "Honesty helps build trust and improves learning."

        elif intent == 3:
            return "You can reduce stress by resting and managing time."

    else:
        return "I can answer only simple educational questions."

# Run chatbot
print("Chatbot started. Type 'bye' to stop.")

while True:
    user = input("You: ")
    print("Bot:", chatbot(user))

    if "bye" in user.lower():
        break
```
